```python
import glob
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cluster
%matplotlib inline
```

```python
data = plt.imread("70.jpg")
plt.imshow(data)
```

```python
print(np.shape(data))
```

```python
X = data.reshape((-1,3))
print(np.shape(X))
```

```python
n_clusters = 6
k_means = sklearn.cluster.KMeans(n_clusters=n_clusters)
k_means.fit(X)
```

```python
# calculo a cual cluster pertenece cada pixel
cluster = k_means.predict(X)
#cluster asigno a cada pixel el lugar del centro de su cluster
X_centered = X.copy()
for i in range(n_clusters):
    ii = cluster==i
    X_centered[ii,:] = np.int_(k_means.cluster_centers_[i])

# devuelvo los datos a las dimensiones originales de la imagen
data_centered = X_centered.reshape((100,100,3))
print(X_centered)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(data)
plt.title("Original")

plt.subplot(1,2,2)
plt.imshow(data_centered)
plt.title("{} colores".format(n_clusters))
```
